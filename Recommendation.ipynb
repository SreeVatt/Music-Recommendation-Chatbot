```python
user_Bolly_df=pd.read_csv('Collabrative Filtering T.csv')
```

```python
starwars_user_rating=user_Bolly_df['Enjoy Enjaami']
starwars_user_rating.head()
```

```python
similar_to_starwars=user_Bolly_df.corrwith(starwars_user_rating)
```

```python
corr_starwars=pd.DataFrame(similar_to_starwars,columns=['Correlation'])
corr_starwars.dropna(inplace=True)
corr_starwars.head()
```

```python
corr_starwars.sort_values('Correlation',ascending=False).head(10)
```
